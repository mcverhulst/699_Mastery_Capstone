```python
import pandas as pd
```

```python
dollar_df = pd.read_csv('Data/inflation_data.csv')
dollar_df.year = dollar_df.year.astype(str)
```

### Edu wages

```python
wages = pd.read_csv('../data/edu_wages.csv')
```

```python
pay = wages.iloc[:, 1:27]
ohe = wages.iloc[:, 27:]
```

```python
pay
```

```python
years = ['1995', '1996', '1997', '1998', '1999',
       '2000', '2001', '2002', '2003', '2004', '2005' ,'2006', '2007', '2008', '2009',
       '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019',
       '2020']

for i in range(len(years)):
    # get series for given year
    x = pay[years[i]]

    # get 2022 dollar figure
    y = dollar_df[dollar_df.year == years[i]]
    y = y['2022_dollar'].values[0]

    # convert series to 2022 dollars
    z = x.apply(lambda x: x * y)

    # change original series to 2022 dollars
    pay[years[i]] = z
```

```python
wages_2022 = wages.copy()
```

```python
wages_2022.iloc[:, 1:27] = pay
wages_2022 = wages_2022.round(0)
```

```python
wages_2022.to_csv('../data/edu_wages_2022.csv', index=False)
```
